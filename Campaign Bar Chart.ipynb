```python
import pandas as pd
import collections as c
from collections import Counter
```

```python
df = pd.read_csv('V3.csv')
```

```python
df.head(5)
```

```python
df.groupby("CAMPAIGN").count()
```

```python
CAMPAIGN = df['CAMPAIGN']
type(CAMPAIGN)

```

```python
Campaign_list = c.Counter(CAMPAIGN).most_common(20)
df = pd.DataFrame(Campaign_list)
Campaign = df[0]
Number = df[1]
```

```python
import matplotlib.pyplot as plt
```
